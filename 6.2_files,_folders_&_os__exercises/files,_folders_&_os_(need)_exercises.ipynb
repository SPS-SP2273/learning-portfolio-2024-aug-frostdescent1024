```python
glob.glob('os-collaboration-exercise-data/*txt', recursive=True)
```

```python
months= {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

def rename_my_file(old_file_name):
    file_month = old_file_name[0:3]
    new_file_name = old_file_name.replace(file_month, months[file_month])
    return new_file_name

directory = glob.glob('os-collaboration-exercise-data/*txt', recursive=True)

for file_names in directory:
    old_file_name = file_names.replace('os-collaboration-exercise-data\\','')
    current_path = os.path.join('.','os-collaboration-exercise-data',old_file_name)
    new_path = os.path.join('.','os-collaboration-exercise-data', rename_my_file(old_file_name))
    shutil.move(current_path, new_path)
#use shutil to move the old name to the new name: use the same directory, but change the name 
```

```python
for file in glob.glob('**\\*cairo*'):
    try:
        path_cairo = os.path.join('.', 'cities', 'Cairo')
        shutil.copy(file, path_cairo)
        print(f'Copied file to {path_cairo}')
    except PermissionError:
        print('this is cities Cairo folder')
```

```python
for city in ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']:
    for file in glob.glob(f'**\\*{city}*'):
        try:
            path_city = os.path.join('.', 'cities', city)
            shutil.copy(file, path_city)
            print(f'Copied file to {path_city}')
        except PermissionError:
            print(f'this is a {city} folder')
```
